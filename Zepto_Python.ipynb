```python
df = raw_data.copy()
```

```python
df.head()
```

### Converting Paisa into Rupees

```python
df['discountedSellingPrice'] = df['discountedSellingPrice'] / 100
```

```python
df['mrp'] = df['mrp'] / 100
```

### Calculate Revenue, DiscountAmount, StockStatus

```python
df['Revenue'] = df['discountedSellingPrice'] * df['availableQuantity']
```

```python
df['DiscountAmount'] = df['mrp'] - df['discountedSellingPrice']  
```

### Establish connection

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="@Nikunj532",
    database="zepto"
)
```

```python
cursor = conn.cursor()
```

```python
sql = """
INSERT INTO products 
(Category, Name, MRP, Discount_Percent, Available_Quantity,
 Discounted_Price, Weight_gms, Out_of_Stock, Quantity, 
 Revenue, Discount_Amount)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
```

### Insert rows from python to MySQL 

```python
for i, row in df.iterrows():
        values = (
            row['Category'],              
            row['name'],
            row['mrp'],
            row['discountPercent'],
            row['availableQuantity'],
            row['discountedSellingPrice'],
            row['weightInGms'],
            row['outOfStock'],
            row['quantity'],
            row['Revenue'],
            row['DiscountAmount']
        )
        cursor.execute(sql, values)
```

```python
conn.commit()
```

```python
cursor.close()
conn.close()

print("Data inserted successfully")
```
